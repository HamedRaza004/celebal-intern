## рҹ“Ӯ STEP 1 вҖ” Load the Dataset

We have two options:
- **Option A:** Load the pre-combined `Combined_dataset.csv`
- **Option B:** Merge all individual CSV files using `glob`

We'll use Option A as primary and show Option B as bonus.

```python
DATASET_PATH = '/home/hamed/Documents/myprojects/archive/Combined_dataset.csv'

df = pd.read_csv(DATASET_PATH, sep=',', on_bad_lines='skip')

print('вң… Loaded Combined_dataset.csv')
print(f'   Shape: {df.shape[0]:,} rows Г— {df.shape[1]} columns')

print(df.columns.tolist())
```

```python
# в”Җв”Җ Option B: Merge ALL individual CSV files (bonus) в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
ARCHIVE_FOLDER = '/home/hamed/Documents/myprojects/archive/'

all_files = glob.glob(os.path.join(ARCHIVE_FOLDER, '*.csv'))
# Exclude the combined file to avoid duplication
individual_files = [f for f in all_files if 'Combined_dataset' not in f]

dfs = []
for filepath in individual_files:
    try:
        temp = pd.read_csv(filepath, sep='\t', on_bad_lines='skip')
        # Add category column from filename
        temp['source_file'] = os.path.basename(filepath).replace('.csv', '')
        dfs.append(temp)
    except Exception as e:
        print(f'вҡ пёҸ  Skipped {os.path.basename(filepath)}: {e}')

df_all = pd.concat(dfs, ignore_index=True)
print(f'вң… Merged {len(dfs)} individual CSV files')
print(f'   Combined shape: {df_all.shape[0]:,} rows Г— {df_all.shape[1]} columns')
print(f'   Categories found: {df_all["source_file"].nunique()}')
```

## рҹ§№ STEP 3 вҖ” Handle Missing Values

Identify which columns have nulls, what percentage, and then fill or drop appropriately.

```python
# Count missing values per column
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
}).query('`Missing Count` > 0').sort_values('Missing %', ascending=False)

print(f'рҹ”Қ Columns with missing values: {len(missing_df)}')
missing_df
```

```python
# Fill missing values
print('рҹ”§ Filling missing values...')

# Numeric columns вҖ” fill with median
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f'   вң… {col}: filled with median ({median_val:.2f})')

# Text columns вҖ” fill with 'Unknown'
text_cols = df.select_dtypes(include=['object']).columns
for col in text_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna('Unknown', inplace=True)
        print(f'   вң… {col}: filled with "Unknown"')

print(f'\nвң… Total missing values remaining: {df.isnull().sum().sum()}')
```

## рҹ”Һ STEP 4 вҖ” Filter Rows & Select Columns

Practical data operations: filtering by conditions and selecting relevant columns.

```python
# Select key columns only
key_cols = ['product_id', 'title', 'rating', 'ratings_count', 
            'initial_price', 'discount', 'final_price', 'category']

# Only keep columns that actually exist
key_cols = [c for c in key_cols if c in df.columns]
df_slim = df[key_cols].copy()

print(f'вң… Selected {len(key_cols)} key columns')
df_slim.head()
```

```python
# Filter: products with initial_price > 500
if 'initial_price' in df.columns:
    df['initial_price'] = pd.to_numeric(df['initial_price'], errors='coerce').fillna(0)
    expensive = df[df['initial_price'] > 500]
    print(f'рҹ’° Products with price > вӮ№500: {len(expensive):,} rows')
    expensive[key_cols].head(10)
```

```python
# Filter: products rated 4.0 and above
if 'rating' in df.columns:
    df['rating'] = pd.to_numeric(df['rating'], errors='coerce').fillna(0)
    top_rated = df[df['rating'] >= 4.0]
    print(f'вӯҗ Products rated 4.0+: {len(top_rated):,} rows')
    top_rated[key_cols].head(10)
```

```python
# Filter: specific category вҖ” tshirts
if 'category' in df.columns:
    tshirts = df[df['category'] == 'tshirts']
    print(f'рҹ‘• T-shirts found: {len(tshirts):,} rows')
    tshirts[key_cols].head(10)
```

## рҹ—‘пёҸ STEP 5 вҖ” Remove Duplicates

Check for and remove duplicate rows to keep the dataset clean.

```python
before = len(df)
dupes = df.duplicated().sum()
print(f'рҹ“Ӣ Total rows before: {before:,}')
print(f'рҹ”Ғ Duplicate rows found: {dupes:,}')

df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

after = len(df)
print(f'вң… Total rows after removing duplicates: {after:,}')
print(f'   Removed: {before - after:,} rows')
```

```python
# Also check duplicates by product_id
if 'product_id' in df.columns:
    id_dupes = df.duplicated(subset=['product_id']).sum()
    print(f'рҹ”Ғ Duplicate product_ids: {id_dupes:,}')
    if id_dupes > 0:
        df.drop_duplicates(subset=['product_id'], keep='first', inplace=True)
        print(f'вң… Kept first occurrence of each product_id')
        print(f'   Final rows: {len(df):,}')
```

## вһ• STEP 6 вҖ” Create Derived Column: `total_amount`

Since this is a product listing dataset (not transactions), there is no `quantity` column.  
We'll simulate quantity with random values (1вҖ“5) and compute `total_amount = final_price Г— quantity`.

```python
# Clean final_price вҖ” remove currency symbols and convert to numeric
if 'final_price' in df.columns:
    df['final_price_clean'] = (
        df['final_price']
        .astype(str)
        .str.replace(r'[вӮ№,$,",\s]', '', regex=True)
        .str.replace(',', '')
    )
    df['final_price_clean'] = pd.to_numeric(df['final_price_clean'], errors='coerce')
    df['final_price_clean'].fillna(df['initial_price'], inplace=True)
    print(f'вң… Cleaned final_price вҖ” sample values:')
    print(df['final_price_clean'].head())
else:
    df['final_price_clean'] = df['initial_price']
```

```python
# Create quantity column (simulated 1-5)
np.random.seed(42)
df['quantity'] = np.random.randint(1, 6, size=len(df))

# Create total_amount = price Г— quantity
df['total_amount'] = df['final_price_clean'] * df['quantity']

print('вң… Created derived columns:')
print(f'   quantity      вҖ” random values 1-5')
print(f'   total_amount  вҖ” final_price Г— quantity')
print()

# Show top 10 highest total_amount products
print('рҹ’° Top 10 Highest Total Amount Products:')
top10_cols = ['title', 'category', 'final_price_clean', 'quantity', 'total_amount']
top10_cols = [c for c in top10_cols if c in df.columns]
df.nlargest(10, 'total_amount')[top10_cols]
```
